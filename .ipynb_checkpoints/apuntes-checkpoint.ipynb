## Que es una funcion?

Concepto de funcion:

- Es una **regla** donde a cada elemento de un conjunto A se le **asigna** un elemento de un conjunto B.

    Una funci√≥n $(f)$ es una relaci√≥n entre un conjunto dado $X$ (llamado dominio ) y otro conjunto de elementos $Y$ (llamado codominio ) de forma que a cada elemento $x$ del dominio le corresponde un √∫nico elemento $f(x)$ del codominio (los que forman el recorrido, tambi√©n llamado rango o √°mbito ).

### Variables Cualitativas (Categoricas)

- Nominales:

    Son las que les asignamos una cualidad. Poseen un valor intrinseco, pero no necesariamente un orden. Nos permite **clasificar objetos**.
    
        Ejemplo (rojo, verde, azul)

- Ordinales:

    Representan un orden. Nos permite **jerarquizar** cosas por sus valores.
        
        Ejemplo (alto, medio, bajo)

- Binarias:

    Solo toman dos valores, usualmente usadas para representar estados. Nos permite escoger entre un estado.
        
        Ejemplo (Existe o no existe, est√° fr√≠o o caliente, uno o cero)

### Variables Cuantitativas (Numericas)

- Discretas:

    Representan un gurpo de datos finitos para tomar ciertos valores. Nos permite **contar** *"pasos"* entre variables separadas
        
        Ejemplo (una persona puede tener 0, 1, 2 o 3 amigos pero nunca 3.5.)

- Continuas:

    Sus valores pueden verse como infinitos al tomar cualquier vlaor dentro de los numeros reales en un rango establecido, podemos usar infinitos decimales para poder expresar sus valores. Nos permite **medir**

        Ejemplo (Medir la estatura de una persona)

### C√≥mo leer las matem√°ticas: Conjuntos
- Los **Diagramas de Venn** son otra forma de representar un conjunto de manera gr√°fica. Los diagramas nos permiten comprender y analizar f√°cilmente los conjuntos.

- Conjunto vaci√≥ ($œï$)
Nota: El conjunto que contenga el elemento $0$ o el $œï$ es no vac√≠o.

- La **uni√≥n** de dos conjuntos $A$ y $B$ es el conjunto formado por todos los elementos de ambos conjuntos sin repetir ninguno y se representa as√≠: $A ‚à™ B = \{x | x ‚àà A √≥ x ‚àà B\}$

- La **intersecci√≥n** de dos conjuntos $A$ y $B$ son el conjunto de elementos que forman parte tanto de $A$ como de $B$, es decir: $A ‚à© B = \{x | x ‚àà A y x ‚àà B\}$

- La diferencia de dos conjuntos $A$ y $B$, es un conjunto formado por todos los elementos de $A$ que no est√°n en $B$ y se escribe as√≠: $A/B=A-B=\{x‚ààA|x‚àâB\}$

- Conjunto universo (Œ©)
Es el conjunto que incluye a todos los elementos y un conjunto $A$ **subconjunto** de $Œ©$. 

- El **complemento** de $A$ son todos aquellos elementos del conjunto universo que no pertenecen a $A$, es decir: $A‚Äô = \{x ‚àà Œ© | x ‚àâ A\}$

- Seg√∫n su n√∫mero de elementos, los conjuntos se pueden clasificar como **finitos o infinitos.**

- La cardinalidad de un conjunto $A$ es la cantidad de elementos diferentes que tiene y se representa con $|A|$

Un conjunto es finito si se conoce su cardinalidad (podemos contar sus elementos) 
Un conjunto es infinito si se desconoce su cardinalidad (no podemos contar sus elementos).

### Funci√≥n lineal

Tiene la forma de $$f(x)=mx + b$$ Donde $m$ y $b$  $\in R$. 


$m$ puede ser calculada por: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

- $m$ se le conoce como **pendiente de la recta**. 
- Y $b$ es **el punto de corte** con el eje $y$. 
- Su **dominio** es $Dom_{f} = (-\infty, \infty)$. 
- Su **imagen** es $Im_{f} = (-\infty, \infty)$

```python
# Ejemplo Funcion Lineal
def graficar(x,y):
    plt.plot(x,y)
    plt.grid()
    plt.axhline(y=0, color='r', linewidth=0.8)
    plt.axvline(x=0, color='r', linewidth=0.8)
    plt.show()
    

N = 100 # N es la resolucion de una funcion

m = 1 # Pendiente de la recta

b = 0 # Punto de corte con el eje Y

# Funcion lineal de primer grado
def f(x):
    return m*x+b

x = np.linspace(-10,10, num=N)

y = f(x)

graficar(x,y)
```

### Funciones polin√≥micas

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una funci√≥n que tiene esta forma se le llama **polinomio de grado $n$**. A los elementos $a$ los llamaremos **coeficientes** donde $a \in R$. 

**Por ejemplo:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$
$$P(x)= 2x^{7} + 0x^6 + 0x^5- x^{4} +0x^3+ 3x^{2} +0x+ 4$$

que es un polinomio de grado 7.

**Ejemplos en codigo:**

```python
# Polinomio Grado IMPAR
def f(x):
    return (2*x**7)-(x**4)+(3*x**2)+4

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
# Polinomio Grado PAR
def f(x):
    return (2*x**8)-(x**4)+(3*x**2)+4

y = f(x)

graficar(x,y)
```

### Funciones potencia

Hay unas funciones que son un caso particular de las funciones polin√≥micas que son las funciones potencia, las cuales tienen la forma:


$$f(x)= x^{a}, a \in R$$ 


**Por ejemplo:**

$$f(x)= x^{2}$$

- El dominio de $f(x)=x^{2}$ es $Dom_{f} = (-\infty, \infty)$. 
- Su imagen es $Im_{f} = [0, \infty)$

**Ejemplos en codigo:**

```python
# Ejemplo Funcion Potencia

def f(x):
    return x**8

y = f(x)

graficar(x,y)
```

## Funciones seccionadas

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la funci√≥n escal√≥n de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
# Ejemplo Escalon de Heaviside
def H(x):
    Y = np.zeros(len(x))

    for idx,x in enumerate(x):
        if x >= 0:
            Y[idx] = 1
    return Y

N = 1000
x = np.linspace(-10,10,N)

y = H(x)

graficar(x,y)
```

```python
# Ejemplo Valor Absoluto
def A(x):
    Y = np.zeros(len(x))

    for idx,x in enumerate(x):
        if x >= 0:
            Y[idx] = x
        else:
            Y[idx] = x*(-1)

    return Y

N = 1000
x = np.linspace(-120,120,N)

y = A(x)

graficar(x,y)
```

## Funciones Compuestas

[Info extra](https://www.fisicalab.com/apartado/composicion-funciones)

**Definici√≥n**

Se define la **funci√≥n compuesta** de dos funciones $f(x)$ y $g(x)$ cualesquiera, y designada por, a la funci√≥n que transforma $x$ en $g[f(x)]:$

$$x \xrightarrow{f} f(x) \xrightarrow{g} g[f(x)] = (g o f)(x)$$

Donde:

: Se lee $f$ **compuesta** con $g$. Es la propia funci√≥n compuesta que permite transformar directamente $x$ en $g[f(x)]$

**Ejemplo Clase**
$$x \xrightarrow{g} g(x) \xrightarrow{f} f[g(x)] = (f o g)(x)$$


```python
# f(g(x))
N = 1000
x = np.linspace(-10,10,num=N)

def g(x):
    return x**2

def f(x):
    return np.sin(x) 

y = g(x)

f_o_g = (f(g(x)))
# f_o_g
graficar(x,f_o_g)
```

## ¬øC√≥mo manipular funciones?


### Desplazamientos verticales y horizontales

Siendo $c$ una **constante** mayor que cero, entonces la gr√°fica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
# Desplazamientos verticales y horizontales
N = 100

def f(x):
    return x**2

c = 5


x = np.linspace(-10,10, num=N)

y2 = f(x) + (c+5)
y3 = f(x) - (c+5)
y4 = f(x - c)  
y5 = f(x + c)  

fig, ax = plt.subplots(2,2, figsize=(18,8))

def graficar2(pos,y,title):
    ax[pos].plot(x,y)
    ax[pos].set_title(title)
    ax[pos].grid()
    ax[pos].axhline(y=0, color='r', linewidth=0.8)
    ax[pos].axvline(x=0, color='r', linewidth=0.8)

graficar2((0,0),y2,'$y = f(x) + (c+5)$')
graficar2((0,1),y3,'$y = f(x) - (c+5)$')
graficar2((1,0),y4,'$y = f(x - c)$')
graficar2((1,1),y5,'$y = f(x + c)$')
plt.tight_layout()
```

### Alargamientos y compresiones

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=c \cdot f(x)$ alarga la gr√°fica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gr√°fica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gr√°fica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gr√°fica horizontelmente en un factor de $c$.

```python
# Ejemplos Alargamientos y Compresion
N = 1000

def f(x):
    return np.sin(x)

c = 2
y = f(x)
y1 = c*f(x)
y2 = (1/c) * f(x)  
y3 = f(c * x) 
y4 = f((1/c) * x) 

graficar(x,y) # Funcion Original


fig, ax = plt.subplots(2,2, figsize=(18,8))

# graficar2(())
graficar2((0,0),y1,'$y = c*f(x)$')
graficar2((0,1),y2,'$y = (1/c) * f(x)$')
graficar2((1,0),y3,'$y = f(c * x)$')
graficar2((1,1),y4,'$y = f((1/c) * x)$')
plt.tight_layout()
```

### Reflexiones

- $y=-f(x)$ refleja la gr√°fica respecto al eje x.
- $y=f(-x)$ refleja la gr√°fica respecto al eje y.

```python
# Ejemplos Reflexion
N = 1000

def f(x):
  return x**3

x = np.linspace(-10,10, num=N)

y1 = f(x)
y2 = -f(x)
y3 = f(-x)

fig, ax = plt.subplots(1,3, figsize=(18,8))
graficar2((0),y1,'$y=f(x)$')
graficar2((1),y2,'$y=f(-x)$')
graficar2((2),y3,'$y=-f(x)$')
fig.tight_layout()
```
